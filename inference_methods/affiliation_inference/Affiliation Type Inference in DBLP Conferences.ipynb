<div class="alert alert-block alert-success">
In order to infer the affiliation type we will first check for some keywords contained in the Affiliation, and for values not appearing here contrast information with University and research centre companies.
</div>

<div class="alert alert-block alert-warning">
    <b>Affiliation classification follows: </b>
    <br>
    
- If univeristy appears (in different languages) -> Academia
- If not, search if appearing in the research centre list  -> Research centre
- If not -> Industry

    <br>
        
    
</div> 

```python
import io
import numpy as np
```

```python
def get_code2(centre):
    
    university = ["University", "College", "Universidad", "Université", "Univ", "Università", "Universitat", 
              "Universidade", "Universiteit", "MIT"]
    research_centre = ["Research", "Centre", "Center", "Institute", "NASA", "Academy", 
                       'Project', 'Instituto', 'Institut', 'Lab', 'Labs', 'Centrum', 'Group', 'CNR']
    
    for item in university:
        if item in centre:
            return "Academia" 
    for item in research_centre:
        if item in centre:
            return "Research Centre"

    return "Industry" 
```

```python
def infer_affiliation_type(df,col):
    df["Code"] = "I"
    codes = []

    for row in df[col]:
        if type(row) is str:
            codes.append(get_code2(row))
        else:
            codes.append(None)

    df.loc[:, 'Code'] = codes
    
    return df
```
